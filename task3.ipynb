```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\Abdul\Downloads\WhatsApp Image 2026-01-19 at 8.40.31 PM.jpeg")
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_lab=cv2.cvtColor(img,cv2.COLOR_BGR2LAB)
img_ycrb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)



```

```python
plt.subplot(1, 4, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Gray Image")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_hsv)  
plt.title("HSV Image")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_lab)
plt.title("LAB Image")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_ycrb)
plt.title("YCrCb Image")
plt.axis('off')
```
