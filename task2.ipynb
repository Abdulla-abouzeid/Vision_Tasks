```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('abdulla.png')
print(img.shape)
print(img.dtype)
print(img.size)
s=img.shape
h,w,c=s[:3]
total_pixel=h*w*c
print(total_pixel)
```

```python
cx,cy=w//2,h//2
c_img=img[cx,cy].copy()
B=img[:,:,0]
G=img[:,:,1]
R=img[:,:,2]

print(f" {B} , G {G} , R {R}")
```

```python
h,w=img.shape[:2]
r=min(w,h)//10
y, x = np.ogrid[:h, :w]
mask=(x-cx)**2 + (y-cy)**2 <= r**2
img[mask]=[255,255,255]
cv2.imshow("abdulla",img)
```
